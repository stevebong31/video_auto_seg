```python
img = cv2.imread(image_list[0])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_c = img.copy()

img_c[..., 1:] = img[..., 1:] * 0.5
img_c[..., 0] = img[..., 0] * 1.1


color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_c],[i],None,[100],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,100])
plt.show()
```
