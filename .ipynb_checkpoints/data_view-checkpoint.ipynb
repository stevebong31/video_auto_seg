```python
import glob, os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels_path = 'C:/Users/HW-Steve/Desktop/DATA/DAVIS/Annotations/480p/'
imgs_path = 'C:/Users/HW-Steve/Desktop/DATA/DAVIS/JPEGImages/480p/'
```

```python
labels_dir_list = os.listdir(labels_path)
imgs_dir_list = os.listdir(imgs_path)
```

```python
imgs_list = glob.glob(os.path.join(imgs_path, imgs_dir_list[0]) +'/*.jpg')
labels_list = glob.glob(os.path.join(labels_path, labels_dir_list[0]) +'/*.png')
```

```python
img = cv2.cvtColor(cv2.imread(imgs_list[0]), cv2.COLOR_BGR2RGB)
lab = cv2.imread(labels_list[0])
```

```python
plt.figure(figsize=(30, 10))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(lab)
plt.subplot(133)
img_lab = img/255 * lab/255
plt.imshow(img_lab)
plt.show
```

```python
img_2 = cv2.cvtColor(cv2.imread(imgs_list[20]), cv2.COLOR_BGR2RGB)
lab_2 = cv2.imread(labels_list[20])
```

```python
plt.figure(figsize=(30, 10))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(img_2)
plt.subplot(133)
plt.imshow(img_2/255 + img_lab)
plt.show
```
